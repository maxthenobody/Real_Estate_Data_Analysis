```python
def bonbubun_allocator(x):
    # x is a string
    try:
        if pd.isna(x) == True:
            return x
        else:
            return str(int(x))
    except:
        return np.nan
```

```python
def buildinggroup_modifier(x):
    # x is a string
    
    try:
        return int(x)
    except:
        return x
```

```python
def add_0(x):
    if pd.isnull(x) == False:
        splitted = x.split('-')
        if len(splitted) == 1:
            return x+'-0'
        else:
            return x
    else:
        return x
```

```python
basedir = './좌표데이터/'
filenames = [f for f in os.listdir(basedir) if (f.endswith('.txt'))]
filenames
```

```python
colnames = ['시군구코드', '출입구일련번호', '법정동코드', '시도명', '시군구명', '읍면동명', '도로명코드', '도로명',
            '지하여부', '건물본번', '건물부번', '건물명', '우편번호', '건물용도분류', '건물군여부', '관할행정동',
            'X좌표', 'Y좌표'
           ]
```

```python
for f in filenames:
    try:
        df = next(pd.read_csv(basedir + f, sep='|', header=None, names=colnames, encoding='ansi', chunksize=1))
    except:
        try:
            df = next(pd.read_csv(basedir + f, sep='|', header=None, names=colnames, chunksize=1))
        except:
            df = next(pd.read_csv(basedir + f, sep='|', header=None, names=colnames, encoding='euc-kr', chunksize=1))
            
    print(df.shape)
    display(df.head())
```

```python
dfs_list = []
for f in tqdm(filenames):
    try:
        df = pd.read_csv(basedir + f, sep='|', header=None, names=colnames, encoding='ansi')
    except:
        try:
            df = pd.read_csv(basedir + f, sep='|', header=None, names=colnames)
        except:
            df = pd.read_csv(basedir + f, sep='|', header=None, names=colnames, encoding='euc-kr')
            
    dfs_list.append(df)
```

```python
concat_df = pd.concat(dfs_list).drop_duplicates().reset_index(drop=True)
print(concat_df.shape)
concat_df.head()
```

```python
concat_df['도로명주소'] = concat_df['시도명'] + ' ' + concat_df['시군구명'] + ' ' + concat_df['도로명'] + ' '\
+ concat_df['건물본번'].apply(str) + '-' + concat_df['건물부번'].apply(str)
coor_df = concat_df[['도로명주소', 'X좌표', 'Y좌표']].drop_duplicates(subset=['도로명주소'])
print(coor_df.shape)
coor_df.head()
```

```python
%%time
coor_df.to_csv('./prepped_data/coor_data_ver_2.csv', index=False)
```
