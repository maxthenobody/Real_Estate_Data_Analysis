```python
import multiprocess as mp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import math
from tqdm import tqdm
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
import datetime
from sklearn.preprocessing import quantile_transform
pd.options.mode.chained_assignment = None  # default='warn'
import gc
```

```python
inter_df = pd.read_csv('./국토교통부_건축물대장_총괄표제부/wbdinfo_baseline_maindata_inter_checpoint_1.csv')
inter_df = inter_df.sort_values(['지번주소', '기준년월']).reset_index(drop=True)
print(inter_df.shape)
inter_df.head()
```

```python
area_df = inter_df[['지번주소', '기준년월', '용적_률_산정_연면적(㎡)']]
area_df['용적연면적'] = area_df.groupby(['지번주소', '기준년월']).transform('max')['용적_률_산정_연면적(㎡)']
area_df = area_df.drop(columns=['용적_률_산정_연면적(㎡)']).drop_duplicates()
area_df.shape
```

```python
area_df.to_csv('./국토교통부_건축물대장_총괄표제부/총괄표제부상_용적연면적.csv', index=False)
```
