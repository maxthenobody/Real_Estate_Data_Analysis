```python
colnames=[
'법정동코드',
'시도명',
'시군구명',
'법정읍면동명',
'법정리명',
'산여부',
'지번본번(번지)',
'지번부번(호)',
'도로명코드',
'도로명',
'지하여부',
'건물본번',
'건물부번',
'건축물대장',
'상세건물명',
'건물관리번호',
'읍면동일련번호',
'행정동코드',
'행정동명',
'우편번호',
'우편일련번호',
'다량배달처명',
'이동사유코드',
'고시일자',
'변동전도로명주소',
'시군구용 건물명',
'공동주택여부',
'기초구역번호',
'상세주소여부',
'비고1',
'비고2']
```

```python
prac_df = next(pd.read_csv(basedir + filenames[0], encoding='ansi', header=None, chunksize=1, sep='|', names=colnames))
print(prac_df.shape)
prac_df
```

```python
addr_df = pd.read_csv(basedir + filenames[0], encoding='ansi', header=None, sep='|', names=colnames)
print(addr_df.shape)
addr_df.head()
```

```python
addr_df['산여부2'] = addr_df['산여부'].apply(lambda x: '산' if x == 1 else '')
```

```python
addr_df['산여부2'].value_counts()
```

```python
addr_df.isna().sum()
```

```python
addr_df['지번주소'] = addr_df['시도명'].apply(lambda x: x.replace(' ','')) + ' ' +\
addr_df['시군구명'].apply(lambda x: x.replace(' ', '')) + ' ' +\
addr_df['법정읍면동명'].apply(lambda x: x.replace(' ', '')) + ' ' +\
addr_df['산여부2'] + addr_df['지번본번(번지)'].apply(str) + '-' + addr_df['지번부번(호)'].apply(str)
```

```python
print(addr_df.shape)
addr_df.head()
```

```python
addr_df['도로명주소'] = addr_df['시도명'].apply(lambda x: x.replace(' ','')) + ' ' +\
addr_df['시군구명'].apply(lambda x: x.replace(' ', '')) + ' ' +\
addr_df['도로명'] + ' ' + addr_df['건물본번'].apply(str) + '-' + addr_df['건물부번'].apply(str)
```

```python
print(addr_df.shape)
addr_df.head()
```

```python
selected_df = addr_df[['지번주소', '도로명주소']].drop_duplicates().reset_index(drop=True)
print(selected_df.shape)
selected_df.head()
```

```python
count = 0
for addr in tqdm(selected_df['지번주소'].unique()):
    addr_addr_df = selected_df[selected_df['지번주소'] == addr]
    if addr_addr_df.shape[0] > 1:
        print(addr)
        display(addr_addr_df)
        count += 1
    
    if count > 5:
        break
```

```python
print(selected_df.shape)
selected_df = selected_df.drop_duplicates(subset=['지번주소']).reset_index(drop=True)
print(selected_df.shape)
selected_df.head()
```

```python
selected_df.to_csv('./prepped_data/addr.csv', index=False)
```
