```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
basedir = './국토교통부_실거래가_공개시스템/상업업무용/'
filenames = [f for f in os.listdir(basedir) if (f.endswith('.csv'))&('(' in f)]
filenames
```

```python
for i, f in enumerate(filenames):
    print(i)
    chunk = pd.read_csv(basedir + f, chunksize=1, encoding='euc-kr', header=15)
    df = next(chunk)
    if '해제사유발생일' in df.columns.tolist():
        df = df.drop(columns=['해제사유발생일'])
        
    print(df.shape)
    display(df.head())
    
    if i == 0:
        first_cols = df.columns
    else:
        print(sum(df.columns == first_cols))
```

```python
dfs_list = []
for i, f in tqdm(enumerate(filenames)):
    df = pd.read_csv(basedir + f, encoding='euc-kr', header=15)
    if '해제사유발생일' in df.columns.tolist():
        df = df.drop(columns=['해제사유발생일'])
    
    dfs_list.append(df)    
```

```python
concat_df = pd.concat(dfs_list).reset_index(drop=True)
print(concat_df.shape)
concat_df.head()
```

```python
concat_df.isna().sum()
```

```python
land_specs_df = pd.read_csv('./prepped_data/land_specs_ver_4.csv')
print(land_specs_df.shape)
land_specs_df.head()
```

```python
land_specs_df['번지'] = land_specs_df['지번주소'].apply(lambda x: x.split(' ')[3])
```

```python
land_specs_df['본번'] = land_specs_df['번지'].apply(lambda x: x.split('-')[0])
```

```python
def masking(x):
    # x is a string
    
    if pd.isna(x) == True:
        return x
    else:
        mask = ''
        for i in range(len(x)-1):
            mask += '*'
        
        return x[0] + mask
```

```python
land_specs_df['본번_masked'] = land_specs_df['본번'].apply(masking)
```

```python
land_specs_df['시군구'] = land_specs_df['지번주소'].apply(lambda x: x.split(' ')[0])\
+ ' ' + land_specs_df['지번주소'].apply(lambda x: x.split(' ')[1])\
+ ' ' + land_specs_df['지번주소'].apply(lambda x: x.split(' ')[2])
```

```python
concat_df.rename(columns={'지번':'본번_masked'}, inplace=True)
```

```python
def doro_condition(x):
    if pd.isna(x) == True:
        return x
    else:
        if '광대' in x:
            return '25m이상'
        elif '중로' in x:
            return '25m미만'
        elif '소로' in x:
            return '12m미만'
        elif '세로' in x:
            return '8m미만'
        else:
            return '-'
```

```python
land_specs_df['도로조건'] = land_specs_df['도로접면'].apply(doro_condition)
```

```python
commercial_df = land_specs_df[(land_specs_df['토지이동상황'] == '상업용')|(land_specs_df['토지이동상황'] == '업무용')].drop(columns=['공시지가', '번지', '본번']).reset_index(drop=True)
print(commercial_df.shape)
commercial_df.head()
```

```python
concat_df.rename(columns={'대지면적':'토지면적', '용도지역':'용도지역명1'}, inplace=True)
```

```python
def land_usage(x):
    # x is a string
    
    if pd.isna(x) == True:
        return x
    else:
        if '업무' in x:
            return '업무용'
        else:
            return '상업용'
```

```python
concat_df['토지이동상황'] = concat_df['건축물주용도'].apply(land_usage)
```

```python
concat_df['토지이동상황'].value_counts()
```

```python
merge_df = concat_df.merge(commercial_df, on=['시군구', '본번_masked', '토지면적', '도로조건', '용도지역명1', '년', '토지이동상황'])
print(merge_df.shape)
merge_df.head()
```

```python
selected_cols = ['지번주소', '건축물주용도', '건축년도', '토지면적', '계약년월', '계약일', '거래금액(만원)']
ls_cols = [col for col in land_specs_df.columns if col not in ['지번주소', '년', '토지면적', '번지', '본번', '본번_masked', '시군구', '도로조건', '공시지가']]
ls_cols
```

```python
selected_df = merge_df[selected_cols + ls_cols]
print(selected_df.shape)
selected_df.head()
```
