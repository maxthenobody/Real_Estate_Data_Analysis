```python
simple_cols = ['법정동명', '지번', '대장구분명', '저촉여부', '용도지역지구명']
land_plan_simple_df = land_plan_df[simple_cols]
land_plan_simple_df.shape
```

```python
land_plan_simple_df.head()
```

```python
def bon_bun_allocator(x):
    splitted = x.split('-')
    return int(splitted[0])

def bu_bun_allocator(x):
    splitted = x.split('-')
    if len(splitted) == 2:
        return int(splitted[1])
    elif len(splitted) == 1:
        return 0
    else:
        print('error')
        return
```

```python
%%time
land_plan_simple_df['본번'] = land_plan_simple_df['지번'].apply(bon_bun_allocator)
land_plan_simple_df['부번'] = land_plan_simple_df['지번'].apply(bu_bun_allocator)
```

```python
land_plan_simple_df.info()
```

```python
land_plan_simple_df.head()
```

```python
land_plan_simple_df['풀주소'] = land_plan_simple_df['법정동명'] + ' ' + land_plan_simple_df['본번'].apply(str) + '-' +\
                                land_plan_simple_df['부번'].apply(str)
land_plan_simple_df.head()
```

```python
land_plan_simple_df.head(20)
```

```python
def land_plan_unstack(df):
    
    df = df[['풀주소', '용도지역지구명']]
    df['이진항'] = 1
    df = df.drop_duplicates()
    df = df.set_index(['풀주소', '용도지역지구명'])
    df = df.unstack(-1, 0)
    
    return df
```

```python
%%time
unstack_df = land_plan_unstack(land_plan_simple_df)
```
