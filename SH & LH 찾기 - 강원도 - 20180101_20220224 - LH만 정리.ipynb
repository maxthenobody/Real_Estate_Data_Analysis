```python
import pandas as pd
import numpy as np
import os
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
pd.options.mode.chained_assignment = None  # default='warn'
plt.rcParams['font.family'] = 'Malgun Gothic'
import datetime
import time
tqdm.tqdm.pandas()
```

```python
new_df = pd.read_excel('./공공주택찾기/강원도/LH_거래사례_20180101_20220224.xlsx')
new_df = new_df.drop(columns=[
    '일괄계약', '도로명', '시군구', '본번', '부번', '건물연식'
])
print(new_df.shape)
new_df.head()
```

```python
concat_df = new_df.copy().sort_values(['지번주소', '건축년도', '부동산유형', '계약날짜', '전용면적(㎡)', '층']).reset_index(drop=True)
print(concat_df.shape)
concat_df.head()
```

```python
concat_df['계약날짜'] = pd.to_datetime(concat_df['계약날짜'])
```

```python
from_2019_df = concat_df.copy()
```

```python
from_2019_df['전용면적(㎡)_최소'] = from_2019_df[['지번주소', '건축년도', '부동산유형', '전용면적(㎡)']].groupby(['지번주소', '건축년도', '부동산유형']).transform('min')['전용면적(㎡)']
from_2019_df['전용면적(㎡)_평균'] = from_2019_df[['지번주소', '건축년도', '부동산유형', '전용면적(㎡)']].groupby(['지번주소', '건축년도', '부동산유형']).transform('mean')['전용면적(㎡)']
from_2019_df['전용면적(㎡)_최대'] = from_2019_df[['지번주소', '건축년도', '부동산유형', '전용면적(㎡)']].groupby(['지번주소', '건축년도', '부동산유형']).transform('max')['전용면적(㎡)']
from_2019_df['전용면적(㎡)_합'] = from_2019_df[['지번주소', '건축년도', '부동산유형', '전용면적(㎡)']].groupby(['지번주소', '건축년도', '부동산유형']).transform('sum')['전용면적(㎡)']
```

```python
from_2019_df['거래금액(만원)_최소'] = from_2019_df[['지번주소', '건축년도', '부동산유형', '거래금액(만원)']].groupby(['지번주소', '건축년도', '부동산유형']).transform('min')['거래금액(만원)']
from_2019_df['거래금액(만원)_평균'] = from_2019_df[['지번주소', '건축년도', '부동산유형', '거래금액(만원)']].groupby(['지번주소', '건축년도', '부동산유형']).transform('mean')['거래금액(만원)']
from_2019_df['거래금액(만원)_최대'] = from_2019_df[['지번주소', '건축년도', '부동산유형', '거래금액(만원)']].groupby(['지번주소', '건축년도', '부동산유형']).transform('max')['거래금액(만원)']
from_2019_df['거래금액(만원)_합'] = from_2019_df[['지번주소', '건축년도', '부동산유형', '거래금액(만원)']].groupby(['지번주소', '건축년도', '부동산유형']).transform('sum')['거래금액(만원)']
```

```python
from_2019_df['층_최대'] = from_2019_df[['지번주소', '건축년도', '부동산유형', '층']].groupby(['지번주소', '건축년도', '부동산유형']).transform('max')['층']
```

```python
from_2019_df['단가(만원/㎡)_최소'] = from_2019_df[['지번주소', '건축년도', '부동산유형', '단가(만원/㎡)']].groupby(['지번주소', '건축년도', '부동산유형']).transform('min')['단가(만원/㎡)']
from_2019_df['단가(만원/㎡)_평균'] = from_2019_df[['지번주소', '건축년도', '부동산유형', '단가(만원/㎡)']].groupby(['지번주소', '건축년도', '부동산유형']).transform('mean')['단가(만원/㎡)']
from_2019_df['단가(만원/㎡)_최대'] = from_2019_df[['지번주소', '건축년도', '부동산유형', '단가(만원/㎡)']].groupby(['지번주소', '건축년도', '부동산유형']).transform('max')['단가(만원/㎡)']
```

```python
from_2019_df.head()
```

```python
summary_df = from_2019_df.drop(columns=['전용면적(㎡)', '거래금액(만원)', '층', '단가(만원/㎡)']).drop_duplicates()
summary_df['구'] = summary_df['지번주소'].apply(lambda x: x.split(' ')[1])
summary_df['동'] = summary_df['지번주소'].apply(lambda x: x.split(' ')[2])
summary_df = summary_df.sort_values(['구', '동', '계약날짜', '지번주소', '부동산유형']).reset_index(drop=True)
summary_df = summary_df.drop(columns=['구', '동'])
print(summary_df.shape)
summary_df.head()
```
