```python
%%time
main_data_df = pd.read_csv('./Prepped Data/main_data_coor_days_20211223.csv')
print(main_data_df.shape)
main_data_df.head()
```

```python
main_data_df['구'] = main_data_df['시군구'].apply(lambda x: x.split(' ')[1])
main_data_df['동'] = main_data_df['시군구'].apply(lambda x: x.split(' ')[2])
```

```python
main_data_df['층'] = main_data_df['층'].astype('Int64')
```

```python
main_data_df['건축년도'] = main_data_df['건축년도'].astype('Int64')
```

```python
main_data_df['계약날짜일수_2006년기준'] = main_data_df['계약날짜일수_2006년기준'].astype('Int64')
```

```python
main_data_df.head()
```

```python
main_data_df['건축년도'].value_counts()
```

```python
main_data_df['건축년도'] = main_data_df['건축년도'].apply(lambda x: '1990이하' if x <= 1990 else str(x))
```

```python
main_data_df.isna().sum()
```

```python
nan_df = main_data_df[main_data_df['건물단지명_키워드'].isna()]
print(nan_df.shape)
nan_df.head()
```

```python
main_data_df['건물단지명_키워드'].fillna('없음', inplace=True)
```

```python
main_data_df.isna().sum()
```

```python
main_data_df = main_data_df.dropna().reset_index(drop=True)
print(main_data_df.shape)
main_data_df.head()
```

```python
main_data_df.dtypes
```

```python
main_data_df['주소_건축년도'] = main_data_df['시군구'] + ' ' + main_data_df['번지'].apply(str) + ' ' + main_data_df['건축년도'].apply(str)
```

```python
main_data_df.head()
```

```python
main_data_df.isna().sum()
```

```python
main_data_df['건물단지명_키워드'].nunique()
```

```python
main_data_df['지목'].value_counts()
```

```python
cat_cols = ['거래유형', '층', '건축년도', '부동산유형', '브랜드', '대장구분명', '구', '동']
```

```python
for col in tqdm.tqdm_notebook(cat_cols):
    unique_cats = main_data_df[col].unique()
    
    # mapping
    cat_map = {}
    for i, v in enumerate(unique_cats):
        cat_map[v] = i
        
    main_data_df[col] = main_data_df[col].map(cat_map)
    
    map_df = pd.DataFrame(cat_map, index=[0])
    map_df.to_excel('./Training/Training1_20211223/cat_col_maps/{}'.format(col+'.xlsx'), index=False)
```

```python
main_data_df.head()
```

```python
training_data_df = main_data_df[['주소_건축년도', '전용면적(㎡)', '월세(만원)', 'x좌표', 'y좌표', '계약날짜일수_2006년기준'] + cat_cols + ['거래/보증금(만원)']]
print(training_data_df.shape)
training_data_df.head()
```
