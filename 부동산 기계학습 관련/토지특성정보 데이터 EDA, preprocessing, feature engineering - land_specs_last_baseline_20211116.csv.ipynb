```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
import numpy as np
import os
import time
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
land_specs_df = pd.read_csv('./Prepped Data/land_specs_baseline_20211116.csv')
print(land_specs_df.shape)
land_specs_df.head()
```

```python
land_specs_df['기준년도'] = land_specs_df['기준년도'].astype('int')
land_specs_df['기준월'] = land_specs_df['기준월'].astype('int')
```

```python
land_specs_df['데이터기준일자'] = pd.to_datetime(land_specs_df['데이터기준일자'])
```

```python
land_specs_df = land_specs_df.sort_values(['법정동명', '지번', '대장구분명', '기준년도', '기준월', '데이터기준일자']).reset_index(drop=True)
print(land_specs_df.shape)
land_specs_df.head()
```

```python
last_df = land_specs_df.drop_duplicates(subset=['법정동명', '지번', '대장구분명'], keep='last').reset_index(drop=True)
print(last_df.shape)
last_df.head()
```
