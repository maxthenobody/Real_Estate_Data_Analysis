```python
def create_numerical_features_per_unique_value(df, colname):
    df = df.copy()
    
    for v in tqdm.tqdm_notebook(df['주_부속_구분'].unique()):
        v = str(v)
        
        v_df = df[df['주_부속_구분'] == v]
        
        colname = str(colname)
        
        df[v+'_'+colname+'_합'] = np.nan
        df[v+'_'+colname+'_평균'] = np.nan
        df[v+'_'+colname+'_최소'] = np.nan
        df[v+'_'+colname+'_중간값'] = np.nan
        df[v+'_'+colname+'_최대'] = np.nan
        
        v_df[v+'_'+colname+'_합'] = v_df[['대지_위치', colname]].groupby(['대지_위치']).transform('sum')[colname]
        v_df[v+'_'+colname+'_평균'] = v_df[['대지_위치', colname]].groupby(['대지_위치']).transform('mean')[colname]
        v_df[v+'_'+colname+'_최소'] = v_df[['대지_위치', colname]].groupby(['대지_위치']).transform('min')[colname]
        v_df[v+'_'+colname+'_중간값'] = v_df[['대지_위치', colname]].groupby(['대지_위치']).transform('median')[colname]
        v_df[v+'_'+colname+'_최대'] = v_df[['대지_위치', colname]].groupby(['대지_위치']).transform('max')[colname]
        
        df.loc[v_df.index, v+'_'+colname+'_합'] = v_df[v+'_'+colname+'_합']
        df.loc[v_df.index, v+'_'+colname+'_평균'] = v_df[v+'_'+colname+'_평균']
        df.loc[v_df.index, v+'_'+colname+'_최소'] = v_df[v+'_'+colname+'_최소']
        df.loc[v_df.index, v+'_'+colname+'_중간값'] = v_df[v+'_'+colname+'_중간값']
        df.loc[v_df.index, v+'_'+colname+'_최대'] = v_df[v+'_'+colname+'_최대']
        
        df[v+'_'+colname+'_합'].fillna(0, inplace=True)
        df[v+'_'+colname+'_평균'].fillna(0, inplace=True)
        df[v+'_'+colname+'_최소'].fillna(0, inplace=True)
        df[v+'_'+colname+'_중간값'].fillna(0, inplace=True)
        df[v+'_'+colname+'_최대'].fillna(0, inplace=True)
    
    df = df.drop(columns=[colname])
    
    return df
```

```python
def find_most_common(x):
    try:
        return x.value_counts().index[0]
    except:
        '없음'
```

```python
def find_second_most_common(x):
    try:
        return x.value_counts().index[1]
    except:
        return '없음'
```

```python
def find_least_common(x):
    try:
        return x.value_counts().index[-1]
    except:
        return '없음'
```

```python
def find_nunique(x):
    try:
        return x.nunique()
    except:
        return '없음'
```

```python
def create_categorical_features_per_unique_value(df, colname):
    df = df.copy()
    
    for v in tqdm.tqdm_notebook(df['주_부속_구분'].unique()):
        v = str(v)
        colname = str(colname)
        
        v_df = df[df['주_부속_구분'] == v]
        
        df[v+'_'+colname+'_most_common'] = np.nan
        df[v+'_'+colname+'_second_most_common'] = np.nan
        df[v+'_'+colname+'_least_common'] = np.nan
        df[v+'_'+colname+'_nunique'] = np.nan
        
        v_df[v+'_'+colname+'_most_common'] = v_df[['대지_위치', colname]].groupby(['대지_위치']).transform(find_most_common)[colname]
        v_df[v+'_'+colname+'_second_most_common'] = v_df[['대지_위치', colname]].groupby(['대지_위치']).transform(find_second_most_common)[colname]
        v_df[v+'_'+colname+'_least_common'] = v_df[['대지_위치', colname]].groupby(['대지_위치']).transform(find_least_common)[colname]
        v_df[v+'_'+colname+'_nunique'] = v_df[['대지_위치', colname]].groupby(['대지_위치']).transform(find_nunique)[colname]
        
        df.loc[v_df.index, v+'_'+colname+'_most_common'] = v_df[v+'_'+colname+'_most_common']
        df.loc[v_df.index, v+'_'+colname+'_second_most_common'] = v_df[v+'_'+colname+'_second_most_common']
        df.loc[v_df.index, v+'_'+colname+'_least_common'] = v_df[v+'_'+colname+'_least_common']
        df.loc[v_df.index, v+'_'+colname+'_nunique'] = v_df[v+'_'+colname+'_nunique']
        
        df[v+'_'+colname+'_most_common'].fillna('없음', inplace=True)
        df[v+'_'+colname+'_second_most_common'].fillna('없음', inplace=True)
        df[v+'_'+colname+'_least_common'].fillna('없음', inplace=True)
        df[v+'_'+colname+'_nunique'].fillna('없음', inplace=True)
    
    df = df.drop(columns=[colname])
    
    return df
```
