```python
%%time
main_data_df = pd.read_csv('./Prepped Data/main_data_coor_days_nonan_20211224.csv', dtype={1:'str', 9:'Int64', 10:'Int64'})
main_data_df['계약날짜'] = pd.to_datetime(main_data_df['계약날짜'])
print(main_data_df.shape)
main_data_df.head()
```

```python
main_data_df['지번주소'] = main_data_df['시군구'] + ' ' + main_data_df['번지']
```

```python
main_data_df = main_data_df.sort_values(['시군구', '본번', '부번', '건축년도', '계약날짜']).reset_index(drop=True)
print(main_data_df.shape)
main_data_df.head()
```

```python
main_data_df['건축년도'].dtype
```

```python
type(main_data_df['건축년도'].max())
```

```python
prac_df = main_data_df[['지번주소', '건축년도']]
print(prac_df.shape)
prac_df.head()
```

```python
prac_df['건축년도_max'] = prac_df.groupby(['지번주소']).transform('max')['건축년도']
```

```python
prac_df.head()
```

```python
new_building_prac_df = prac_df[prac_df['건축년도'] == prac_df['건축년도_max']]
print(new_building_prac_df.shape)
new_building_prac_df.head()
```

```python
new_building_df = main_data_df.loc[new_building_prac_df.index, :].reset_index(drop=True)
print(new_building_df.shape)
new_building_df.head()
```

```python
land_specs_df = pd.read_csv('./Prepped Data/land_specs_last_baseline_20211116.csv')
print(land_specs_df.shape)
land_specs_df.head()
```

```python
new_building_df = new_building_df.rename(columns={'번지':'지번'})
```

```python
def remove_spaces(x):
    # x is a string
    
    if pd.isna(x) == True:
        return x
    
    x = str(x)
    
    if x[0] == ' ':
        x = x[1:]
    
    if x[-1] == ' ':
        x = x[:-1]
    
    return x
```

```python
land_specs_df = land_specs_df.rename(columns={'법정동명':'시군구'})
```

```python
new_building_df['시군구'] = new_building_df['시군구'].apply(remove_spaces)
land_specs_df['시군구'] = land_specs_df['시군구'].apply(remove_spaces)
```

```python
land_specs_df = land_specs_df.drop(columns=['공시지가'])
```

```python
land_specs_df = land_specs_df.drop(columns=['기준년도', '기준월'])
```

```python
new_building_df['대장구분명'].value_counts()
```

```python
land_specs_df['대장구분명'].value_counts()
```

```python
land_specs_df['대장구분명'] = land_specs_df['대장구분명'].apply(lambda x: '산' if x == '산' else '일반')
```

```python
land_specs_df['데이터기준일자'] = pd.to_datetime(land_specs_df['데이터기준일자'])
```

```python
land_specs_df = land_specs_df.sort_values(['시군구', '대장구분명', '지번', '데이터기준일자']).reset_index(drop=True)
```

```python
land_specs_df.shape
```

```python
land_specs_df = land_specs_df.drop_duplicates(subset=['시군구', '대장구분명', '지번'], keep='last')
print(land_specs_df.shape)
```

```python
merge_df = new_building_df.merge(land_specs_df, on=['시군구', '대장구분명', '지번'])
print(merge_df.shape)
merge_df.head()
```

```python
%%time
merge_df.to_csv('./Prepped Data/main_data_new_building_merge_land_specs_20211228.csv', index=False)
```
