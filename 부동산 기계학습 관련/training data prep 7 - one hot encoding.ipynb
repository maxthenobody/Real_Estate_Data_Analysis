```python
import pandas as pd
import numpy as np
import os
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
pd.options.mode.chained_assignment = None  # default='warn'
plt.rcParams['font.family'] = 'Malgun Gothic'
font = {'size': 16}
matplotlib.rc('font', **font)
tqdm.tqdm.pandas()
```

```python
%%time
training_data_df = pd.read_csv('./Prepped Data/training_data6_20220207.csv')
print(training_data_df.shape)
training_data_df.head()
```

```python
os.listdir('./Training/Training6_20220207/cat_col_maps/')
```

```python
basedir = './Training/Training6_20220207/cat_col_maps/'
```

```python
filenames = os.listdir(basedir)
```

```python
filenames
```

```python
for file in filenames:
    map_df = pd.read_excel(basedir + file)
    print(map_df.shape)
    display(map_df)
    
    col = file.replace('.xlsx', '')
    
    
    
    reverse_map_dict = {}
    
    for i in range(map_df.shape[1]):
        v = map_df.columns[i]
        idx = map_df.iloc[0, i]
        
        reverse_map_dict[idx] = v
    
    print(reverse_map_dict)
    
    training_data_df[col] = training_data_df[col].map(reverse_map_dict)
```

```python
training_data_df.head()
```

```python
training_data_df = training_data_df.drop(columns=['지목'])
```

```python
training_data_df.head()
```

```python
training_data_df['대장구분명'].value_counts()
```

```python
training_data_df = training_data_df.drop(columns=['대장구분명'])
```

```python
training_data_df.head()
```

```python
cat_cols = ['부동산유형', '브랜드', '구', '동', '일괄계약']
```

```python
cat_cols
```

```python
for col in tqdm.tqdm_notebook(cat_cols):
    uniques = training_data_df[col].unique()
    
    for v in uniques:
        training_data_df[col + '_' + v] = training_data_df[col].apply(lambda x: 1 if x == v else 0)
```

```python
training_data_df = training_data_df.drop(columns=cat_cols)
print(training_data_df.shape)
training_data_df.head()
```

```python
%%time
training_data_df.to_csv('./Prepped Data/training_data7_one_hot_encoding_20220207.csv', index=False)
```
