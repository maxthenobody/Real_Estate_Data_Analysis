```python
basedir = './서울시 개별공시지가 정보/'
filenames = [f for f in os.listdir(basedir) if '.csv' in f]
filenames
```

```python
dfs_list = []
for filename in tqdm.tqdm_notebook(filenames):
    df = pd.read_csv(basedir + filename, encoding='euc-kr')
    dfs_list.append(df)
```

```python
official_price_df = pd.concat(dfs_list)
```

```python
official_price_df['토지코드'] = official_price_df['토지코드'].apply(lambda x: int(x))
```

```python
official_price_df['토지코드'] = official_price_df['토지코드'].astype('Int64')
```

```python
official_price_df['본번_토지코드'] = official_price_df['토지코드'].apply(lambda x: int(str(x)[-8:-4]))
official_price_df['부번_토지코드'] = official_price_df['토지코드'].apply(lambda x: int(str(x)[-4:]))
```

```python
print(official_price_df.shape)
official_price_df.head()
```

```python
cols_to_drop = ['토지코드', '시군구코드', '법정동코드', '필지구분코드']
```

```python
official_price_df = official_price_df.drop(columns=cols_to_drop)
print(official_price_df.shape)
official_price_df.head()
```

```python
official_price_df = official_price_df.drop(columns=['시도명'])
```

```python
official_price_df.head()
```

```python
official_price_df = official_price_df.sort_values(['시군구명', '법정동명', '본번', '부번', '필지구분명', '기준년도', '기준년월']).reset_index(drop=True)
print(official_price_df.shape)
official_price_df.head()
```

```python
official_price_df['필지구분명'].value_counts()
```

```python
official_price_df['대장구분명'] = official_price_df['필지구분명'].apply(lambda x: '산' if x == '임야' else '일반')
```

```python
official_price_df = official_price_df[
    (official_price_df['필지구분명'] == '토지')
    |(official_price_df['필지구분명'] == '임야')
]
```

```python
official_price_df = official_price_df.reset_index(drop=True)
```

```python
official_price_df = official_price_df.rename(columns={'본번':'본번_original', '부번':'부번_original',
                                                      '본번_토지코드':'본번', '부번_토지코드':'부번'
                                                     })
```

```python
official_price_df.head()
```

```python
official_price_df = official_price_df.drop(columns=['본번_original', '부번_original'])
print(official_price_df.shape)
```

```python
official_price_df.head()
```

```python
official_price_df = official_price_df.sort_values(['시군구명', '법정동명', '대장구분명', '본번', '부번', '기준년도', '기준년월']).reset_index(drop=True)
print(official_price_df.shape)
official_price_df.head()
```

```python
official_price_df = official_price_df.drop_duplicates(subset=['시군구명', '법정동명', '대장구분명', '본번', '부번', '기준년도'], keep='last')
print(official_price_df.shape)
official_price_df.head()
```
