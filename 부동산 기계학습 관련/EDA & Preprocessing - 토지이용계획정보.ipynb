```python
def remove_spaces(x):
    # x is a string
    
    if pd.isna(x) == True:
        return x
    else:
        x = str(x)
        
        while x.startswith(' '):
            x = x[1:]
        while x.endswith(' '):
            x = x[:-1]
        
        return x
```

```python
concat_df['법정동명'] = concat_df['법정동명'].apply(remove_spaces)
```
