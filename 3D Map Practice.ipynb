```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
import numpy as np
import os
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python
merge_df = pd.read_csv('./prepped_data/addr_coor_df_20210917.csv')
```

```python
print(merge_df.shape)
merge_df.head()
```

```python
sample_df = merge_df.sample(frac=0.01)
```

```python
coor_x = sample_df['X좌표'].values
```

```python
coor_x.shape
```

```python
coor_y = sample_df['Y좌표'].values
```

```python
coor_y.shape
```

```python
price_z = sample_df['공시지가'].values / 10000
price_z.shape
```

```python
%matplotlib notebook
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,10))

# setting color bar
color_map = cm.ScalarMappable(cmap=cm.Greens_r)
color_map.set_array(price_z)

# Plot the surface.
img = ax.scatter(coor_x, coor_y, price_z, s=0.1)
plt.colorbar(color_map)

# Add a color bar which maps values to colors.
#plt.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
