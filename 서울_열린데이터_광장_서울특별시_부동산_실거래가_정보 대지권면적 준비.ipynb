```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
basedir = './서울_열린데이터_광장_서울특별시_부동산_실거래가_정보/'
filenames = [f for f in os.listdir(basedir) if (f.endswith('.csv'))&('실거래가' in f)]
filenames
```

```python
for i, f in enumerate(filenames):
    print(i)
    chunk = pd.read_csv(basedir + f, chunksize=1, encoding='euc-kr')
    df = next(chunk)
    print(df.shape)
    
    if i == 0:
        first_cols = df.columns
    elif i == 13:
        df = df.rename(columns={'실거래가아이디"':'실거래가아이디'})
        print(sum(df.columns == first_cols))
    else:
        print(sum(df.columns == first_cols))
```

```python
dfs_list = []
for i, f in tqdm(enumerate(filenames)):
    df = pd.read_csv(basedir + f, encoding='euc-kr')
    if i == 13:
        df = df.rename(columns={'실거래가아이디"':'실거래가아이디'})
    
    dfs_list.append(df)
```

```python
concat_df = pd.concat(dfs_list).reset_index(drop=True)
print(concat_df.shape)
concat_df.head()
```

```python
concat_df = concat_df.dropna(subset=['대지권면적']).reset_index(drop=True)
```

```python
concat_df['건물주용도'].value_counts()
```

```python
dandok_df = concat_df[concat_df['건물주용도'] == '단독주택']
```

```python
dandok_df.head()
```

```python
def isintable(x):
    if pd.isna(x) == True:
        return x
    else:
        x = str(x)
        try:
            int(x)
            return 1
        except:
            return np.nan
```

```python
concat_df['intable'] = concat_df['지번코드'].apply(isintable)
```

```python
concat_df.shape
```

```python
concat_df = concat_df.dropna(subset=['intable'])
```

```python
concat_df['지번주소'] = '서울특별시 ' + concat_df['자치구명'] + ' ' + concat_df['법정동명'] + ' ' +\
concat_df['지번코드'].apply(lambda x: str(int(str(x)[-8:-4]))) + '-' +\
concat_df['지번코드'].apply(lambda x: str(int(str(x)[-4:])))
```

```python
concat_df['지번주소'].unique()[:50]
```

```python
concat_df.head()
```

```python
selected_df = concat_df[['지번주소', '건축년도', '건물면적', '대지권면적']].dropna().drop_duplicates().reset_index(drop=True)
selected_df = selected_df[selected_df['대지권면적'] > 0]
print(selected_df.shape)
selected_df.head()
```

```python
selected_df.to_csv(basedir + '실거래_대지권면적.csv', index=False)
```
