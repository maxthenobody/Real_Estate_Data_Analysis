# 역세권 청년주택사업 관련

```python
def 역세권_청년주택_가능여부_확인():
    면적요건 = '불충족'
    역세권요건 = '불충족'
    용도지역요건 = '불충족'
    노후도요건 = '불충족'
    
    역세권_청년주택_가능_용도지역 = ['2종일반', '2종일반(7층이하)', '3종일반', '준주거', '준공업', '근린상업', '일반상업']
    
    if 사업부지_면적 >= 1000:
        면적요건 = '충족'
    
    if 청년주택_역세권여부 == '예':
        역세권요건 = '충족'
    
    if 현재_용도지역 in 역세권_청년주택_가능_용도지역:
        용도지역요건 = '충족'
    
    if (현재건물_종류 == '비주택') | ((현재건물_종류 == '주택') & (현재건물_연식 >= 20)):
        노후도요건 = '충족'
    
    if (면적요건 == '충족') & (역세권요건 == '충족') & (용도지역요건 == '충족') & (노후도요건 == '충족'):
        print('역세권 청년주택 사업대상지 요건을 모두 충족함')
        
        return  True
    
    else:
        print('역세권 청년주택 사업대상지 요건을 모두 충족하지 못함.')
        print('이하 불충족 요건들:')
        if 면적요건 == '불충족':
            print('면적요건')
        if 역세권요건 == '불충족':
            print('역세권요건')
        if 용도지역요건 == '불충족':
            print('용도지역요건')
        if 노후도요건 == '불충족':
            print('노후도요건')
            
        return False
```

```python
def 역세권_청년주택_용도지역_변경():
    준주거_상향가능여부 = False
    일반_또는_근린상업_상향가능여부 = False
    
    준주거_상향가능여부 = 역세권_청년주택_준주거_상향가능여부_확인()
    
    if 준주거_상향가능여부 == True:
        print('역세권 청년주택 준주거로 종상향 가능')
        
    일반_또는_근린상업_상향가능여부 = 역세권_청년주택_일반_또는_근린상업_상향가능여부_확인()
    
    if 일반_또는_근린상업_상향가능여부 == True:
        print('역세권 청년주택 일반 또는 근린상업으로 종상향 가능')
        
    if ((준주거_상향가능여부 == False) & (일반_또는_근린상업_상향가능여부 == False)):
        print('역세권 청년주택 종상향 불가능')
```

```python
def 역세권_청년주택_준주거_상향가능여부_확인():
    준주거_상향가능_현재_용도지역 = ['2종일반', '2종일반(7층이하)', '3종일반']
    가 = False
    나 = False
    다 = False
    라 = False
    
    if (현재_용도지역 in 준주거_상향가능_현재_용도지역) & ((접한_최고_용도지역 == '준주거')|('상업' in 접한_최고_용도지역)):
        가 = True
    if (현재_용도지역 in 준주거_상향가능_현재_용도지역) & (가장_넓은_접도로_넓이 >= 20) & (간선도로에_접함 == '예'):
        나 = True
    if (역세권내_준주거_상업지역_여부 != '해당없음') | (중심지여부 != '해당없음'):
        다 = True
    if 사업부지_면적 >= 2000:
        라 = True
    
    if 가 == True:
        print('7페이지 가)에 해당')
    if 나 == True:
        print('7페이지 나)에 해당')
    if 다 == True:
        print('7페이지 다)에 해당')
    if 라 == True:
        print('7페이지 라)에 해당')
        
    if (가 == True) | (나 == True) | (다 == True) | (라 == True):
        return True
    
    return False
```

```python
def 역세권_청년주택_일반_또는_근린상업_상향가능여부_확인():
    가 = False
    나 = False
    다 = False
    라 = False
    마 = False
    
    노선상업_사업대상지 = ['2종일반', '2종일반(7층이하)', '3종일반', '준주거']
    라_사업대상지 = ['2종일반', '2종일반(7층이하)']
    
    if ((현재_용도지역 == '3종일반') | (현재_용도지역 == '준주거')) & ('상업' in 접한_최고_용도지역):
        가 = True
    if (((현재_용도지역 in 노선상업_사업대상지) & (노선상업지역_비율 >= 0.2))
        | ((현재_용도지역 in 노선상업_사업대상지) & (노선상업지역_비율 > 0) & (사업부지_면적 >= 2000))):
        나 = True
    if ((현재_용도지역 in 노선상업_사업대상지)
        & ('상업' in 역세권내_준주거_상업지역_여부)
        & (가장_넓은_접도로_넓이 >= 20)
        & (간선도로에_접함 == '예')):
        다 = True
    if (현재_용도지역 in 라_사업대상지) & ('상업' in 바로_접한_최고_용도지역):
        라 = True
    if (현재_용도지역 == '근린상업') & (가장_넓은_접도로_넓이 >= 20) & (간선도로에_접함 == '예'):
        마 = True
        
    if 가 == True:
        print('8페이지 가)에 해당')
    if 나 == True:
        print('8페이지 나)에 해당')
    if 다 == True:
        print('8페이지 다)에 해당')
    if 라 == True:
        print('8페이지 라)에 해당')
    if 마 == True:
        print('8페이지 마)에 해당')
        
    if (가 == True) | (나 == True) | (다 == True) | (라 == True) | (마 == True):
        return True
    
    return False
```

# 역세권 공공임대주택 관련

```python
def 역세권_공공임대주택_사업대상지_기준():
    # 불가 지역 완화 가능: 5페이지 2-1-3 참고
    if 도시자연공원구역_근린공원_자연경관지구_최고고도지구_및_전용주거지역_등과_접한_경우 == True:
        print('역세권 공공임대주택 불가 지역')
        return False
    
    역세권요건 = False
    면적요건 = False
    노후도요건 = False # 지구단위로 진행이 쉬움. 5페이지 2-1-4 가. 참고.
    용도지역요건 = False
    역세권_공공임대주택_용도지역_목록 = ['2종일반', '2종일반(7층이하)', '3종일반']
    
    역세권_1차 = False
    역세권_2차 = False
    
    if ((최근접_역입구_직선거리 <= 400) | (최근접_역중심_직선거리 <= 400)):
        역세권_1차 = True
    if ((최근접_역입구_직선거리 <= 600) | (최근접_역중심_직선거리 <= 600)):
        역세권_2차 = True
    
    
    if 역세권_1차 == True:
        역세권요건 = True
    if 사업부지_면적 >= 3000:
        면적요건 = True
    if 현재건물_연식 >= 20:
        노후도요건 = True
    if 현재_용도지역 in 역세권_공공임대주택_용도지역_목록:
        용도지역요건 = True
    
    if ((역세권요건 == True) & (면적요건 == True) & (노후도요건 == True) & (용도지역요건 == True)):
        print('역세권 공공임대주택 사업대상지 요건을 모두 충족함.')
        
        return True
    
    else:
        print('역세권 공공임대주택 사업대상지 요건을 모두 충족하지 못함.')
        print('이하 불충족 요건들:')
        if 역세권요건 == False:
            print('역세권요건')
        if 면적요건 == False:
            print('면적요건')
        if 노후도요건 == False:
            print('노후도요건')
        if 용도지역요건 == False:
            print('용도지역요건')
            
        return False
```
