```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
import os
from tqdm import tqdm
import numpy as np
```

```python
%%time
main_data_df = pd.read_csv('./prepped_data/main_data_ver_6.csv')
print(main_data_df.shape)
main_data_df.head()
```

```python
addr_df = pd.read_csv('./prepped_data/addr_data_ver_1.csv')
print(addr_df.shape)
addr_df.head()
```

```python
merge_df = main_data_df.merge(addr_df, on=['지번주소'], how='left')
print(merge_df.shape)
merge_df.head()
```

```python
nan_df = merge_df[merge_df['도로명주소'].isna()]
print(nan_df.shape)
nan_df.head()
```

```python
nan_df['지번주소'].nunique()
```

```python
nan_df['주택유형'].value_counts()
```

```python
addr_to_download_df = nan_df[['지번주소']].drop_duplicates()
print(addr_to_download_df.shape)
addr_to_download_df.head()
```

```python
addr_to_download_df.to_csv('./prepped_data/addr_to_download.csv', index=False)
```
