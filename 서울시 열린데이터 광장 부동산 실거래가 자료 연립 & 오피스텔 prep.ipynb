```python
import multiprocess as mp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import math
from tqdm import tqdm
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
import datetime
from sklearn.preprocessing import quantile_transform
pd.options.mode.chained_assignment = None  # default='warn'
import gc
import requests
from urllib.parse import urlparse
```

```python
basedir = './서울_열린데이터_광장_서울특별시_부동산_실거래가_정보/'
filenames = [f for f in os.listdir(basedir) if f.endswith('.csv')]
filenames
```

```python
for i, f in tqdm(enumerate(filenames)):
    chunk = pd.read_csv(basedir + f, encoding='euc-kr', chunksize=1)
    df = next(chunk)
    print(df.shape)
    display(df.head())
    if i == 0:
        first_cols = df.columns
    else:
        print(df.columns == first_cols)
```

```python
dfs_list = []
for f in tqdm(filenames):
    df = pd.read_csv(basedir + f, encoding='euc-kr')
    year = df['신고년도'].iloc[0]
    print(year)
    print(type(year))
    if year == 2019:
        df.rename(columns={'실거래가아이디"':'실거래가아이디'}, inplace=True)
    dfs_list.append(df)
```

```python
concat_df = pd.concat(dfs_list).reset_index(drop=True)
```

```python
selected_df = concat_df[(concat_df['건물주용도'].str.contains('연립'))|(concat_df['건물주용도'].str.contains('오피'))]
print(selected_df.shape)
```

```python
def id_3rd(x):
    # x is a stirng
    if pd.isna(x) == True:
        return x
    else:
        return x.split('-')[2]
```

```python
concat_df['실거래가_3번째'] = concat_df['실거래가아이디'].apply(id_3rd)
```

```python
concat_df['실거래가_3번째'].value_counts()
```

```python
def id_5th(x):
    # x is a string
    if pd.isna(x) == True:
        return x
    else:
        return x.split('-')[4]
```

```python
concat_df['실거래가_5번째'] = concat_df['실거래가아이디'].apply(id_5th)
```

```python
concat_df['실거래가_5번째'].value_counts()
```
