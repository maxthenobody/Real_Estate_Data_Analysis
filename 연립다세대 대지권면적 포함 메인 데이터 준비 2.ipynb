```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

```python
basedir = './국토교통부_실거래가_공개시스템/연립다세대/original/'
filenames = [f for f in os.listdir(basedir) if (f.endswith('.csv'))]#&('실거래가' in f)]
filenames
```

```python
for i, f in enumerate(filenames):
    print(i)
    chunk = pd.read_csv(basedir + f, chunksize=1, encoding='euc-kr', header=15)
    df = next(chunk)
    print(df.shape)
    if df.shape[1] != 13:
        df = df.drop(columns=['해제사유발생일'])
        
    display(df.head())
    
    if i == 0:
        first_cols = df.columns
    else:
        print(df.columns == first_cols)
```

```python
dfs_list = []
for i, f in tqdm(enumerate(filenames)):
    df = pd.read_csv(basedir + f, encoding='euc-kr', header=15)
    if df.shape[1] != 13:
        df = df.drop(columns=['해제사유발생일'])
    
    dfs_list.append(df)    
```

```python
concat_df = pd.concat(dfs_list).reset_index(drop=True)
print(concat_df.shape)
concat_df.head()
```

```python
concat_df.rename(columns={'전용면적(㎡)':'전용면적', '대지권면적(㎡)':'대지권면적'}, inplace=True)
```

```python
concat_df.to_csv(basedir + '연립다세대_매매_20130101_20210322.csv', index=False)
```
