```python
import pandas as pd
import numpy as np
import os
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
pd.options.mode.chained_assignment = None  # default='warn'
plt.rcParams['font.family'] = 'Malgun Gothic'
import datetime
```

```python
%%time
main_data_df = pd.read_csv('./prepped_data/메인_데이터_20060101_20210416_ver_2.csv')
print(main_data_df.shape)
main_data_df.head()
```

```python
coor_df = pd.read_csv('./prepped_data/addr_coor_data_ver_1.csv')
print(coor_df.shape)
coor_df.head()
```

```python
main_data_df.isna().sum()
```

```python
doro_nan_df = main_data_df[main_data_df['도로명'] == ' ']
print(doro_nan_df.shape)
doro_nan_df.head()
```

```python
main_data_df.loc[doro_nan_df.index, '도로명'] = np.nan
```

```python
main_data_df = main_data_df.dropna(subset=['도로명']).reset_index(drop=True)
print(main_data_df.shape)
```

```python
main_data_df.isna().sum()
```

```python
def landnum_modifier(x):
    # x is a string
    
    if pd.isna(x) == True:
        return x
    else:
        x = str(x)        
        splitted = x.split('-')
        if len(splitted) == 1:
            return x + '-0'
        else:
            return x
```

```python
main_data_df['도로명'] = main_data_df['도로명'].apply(landnum_modifier)
```

```python
main_data_df['도로명'] = main_data_df['지번주소'].apply(lambda x: x.split(' ')[0]) + ' '\
+ main_data_df['지번주소'].apply(lambda x: x.split(' ')[1]) + ' ' + main_data_df['도로명']
```

```python
main_data_df.rename(columns={'도로명': '도로명주소'}, inplace=True)
```

```python
main_data_df.head()
```

```python
temp_coor_df = coor_df.drop(columns=['지번주소']).drop_duplicates(subset=['도로명주소'])
print(temp_coor_df.shape)
temp_coor_df.head()
```

```python
merge_df = main_data_df.merge(temp_coor_df, on=['도로명주소'])
print(merge_df.shape)
merge_df.head()
```

```python
merge_df['구'] = merge_df['지번주소'].apply(lambda x: x.split(' ')[1])
merge_df['동'] = merge_df['지번주소'].apply(lambda x: x.split(' ')[2])
```

```python
merge_df.head()
```

```python
%%time
merge_df.to_csv('./prepped_data/메인_데이터_20060101_20210416_ver_3.csv', index=False)
```
