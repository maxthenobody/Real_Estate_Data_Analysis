```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
basedir = './지번주소도로명주소/주소/'
filenames = [f for f in os.listdir(basedir) if f.endswith('.txt')]
filenames
```

```python
jibun_filenames = [f for f in filenames if '지번' in f]
jibun_filenames
```

```python
juso_filenames = [f for f in filenames if '주소' in f]
juso_filenames
```

```python
doro_filenames = [f for f in filenames if '도로명' in f]
doro_filenames
```

```python
jibun_colnames = ['관리번호', '일련번호', '법정동코드', '시도명', '시군구명', '법정읍면동명', '법정리명', '산여부', '지번본번',
                  '지번부번', '대표여부'
                 ]
```

```python
dfs_list = []
for f in tqdm(jibun_filenames):
    df = pd.read_csv(basedir + f, sep='|', encoding='ansi', header=None, names=jibun_colnames)
    dfs_list.append(df)
```

```python
jibun_df = pd.concat(dfs_list).drop_duplicates().reset_index(drop=True)
print(jibun_df.shape)
jibun_df.head()
```

```python
jibun_df['관리번호'].nunique()
```

```python
juso_colnames = ['관리번호', '도로명코드', '읍면동일련번호', '지하여부', '건물본번', '건물부번', '기초구역번호',
                 '변경사유코드', '고시일자', '변경전도로명주소', '상세주소부여 여부'
                ]
```

```python
dfs_list = []
for f in tqdm(juso_filenames):
    df = pd.read_csv(basedir + f, sep='|', encoding='ansi', header=None, names=juso_colnames)
    dfs_list.append(df)
```

```python
juso_df = pd.concat(dfs_list).drop_duplicates().reset_index(drop=True)
print(juso_df.shape)
juso_df.head()
```

```python
juso_df['관리번호'].nunique()
```

```python
doro_colnames = ['도로명코드', '도로명', '도로명로마자', '읍면동일련번호', '시도명', '시도 로마자', '시군구명',
                 '시군구 로마자', '읍면동명', '읍면동 로마자', '읍면동구분', '읍면동코드', '사용여부', '변경사유',
                 '변경이력정보', '고시일자', '말소일자'
                ]
```

```python
dfs_list = []
for f in tqdm(doro_filenames):
    df = pd.read_csv(basedir + f, sep='|', encoding='ansi', header=None, names=doro_colnames)
    dfs_list.append(df)
```

```python
doro_df = pd.concat(dfs_list).drop_duplicates().reset_index(drop=True)
print(doro_df.shape)
doro_df.head()
```

```python
prac_jibun_df = jibun_df[jibun_df['관리번호'] == '1111010100100030065030970']
print(prac_jibun_df.shape)
prac_jibun_df
```

```python
prac_juso_df = juso_df[juso_df['관리번호'] == '1111010100100030065030970']
print(prac_juso_df.shape)
prac_juso_df
```

```python
jibun_df = jibun_df.drop(columns=['법정리명'])
```

```python
juso_df = juso_df.drop(columns=['변경사유코드', '고시일자', '변경전도로명주소'])
```

```python
merge_df = jibun_df.merge(juso_df, on=['관리번호'])
print(merge_df.shape)
merge_df.head()
```

```python
prac_df = merge_df[merge_df['도로명코드'] == 111102005001]
print(prac_df.shape)
prac_df
```

```python
merge2_df = merge_df.merge(doro_df, on=['도로명코드', '읍면동일련번호', '시도명', '시군구명'])
print(merge2_df.shape)
```

```python
merge2_df.head()
```

```python
merge2_df = merge2_df.drop(columns=[col for col in merge2_df.columns if '로마자' in col])
print(merge2_df.shape)
merge2_df.head()
```

```python
merge2_df = merge2_df.drop(columns=['변경사유', '변경이력정보', '고시일자', '말소일자'])
print(merge2_df.shape)
merge2_df.head()
```

```python
merge2_df.to_csv(basedir + '지번_도로명_주소.csv', index=False)
```
