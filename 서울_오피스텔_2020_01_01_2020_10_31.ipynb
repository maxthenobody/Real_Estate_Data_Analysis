```python
def bon_bun_allocator(x):
    splitted = x.split('-')
    return int(splitted[0])
```

```python
def bu_bun_allocator(x):
    splitted = x.split('-')
    if len(splitted) == 2:
        return int(splitted[1])
    elif len(splitted) == 1:
        return 0
    else:
        print('error')
        return
```

```python
%%time
land_plan_simple_df['본번'] = land_plan_simple_df['지번'].apply(bon_bun_allocator)
```

```python
land_plan_simple_df.info()
```

```python
land_plan_simple_df.head()
```

```python
%%time
land_plan_simple_df['부번'] = land_plan_simple_df['지번'].apply(bu_bun_allocator)
```
