```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
import os
from tqdm import tqdm
import numpy as np
```

```python
%%time
main_data_df = pd.read_csv('./prepped_data/main_data_ver_6.csv')
print(main_data_df.shape)
main_data_df.head()
```

```python
addr_df = pd.read_csv('./prepped_data/addr_data_ver_1.csv')
print(addr_df.shape)
addr_df.head()
```

```python
addr_df2 = pd.read_csv('./prepped_data/additional_addr_1.csv')
print(addr_df2.shape)
addr_df2.head()
```

```python
addr_df2 = addr_df2.dropna()
print(addr_df2.shape)
```

```python
addr_df = pd.concat([addr_df, addr_df2]).drop_duplicates(subset=['지번주소']).reset_index(drop=True)
print(addr_df.shape)
addr_df.head()
```

```python
merge_df = main_data_df.merge(addr_df, on=['지번주소'])
print(merge_df.shape)
merge_df.head()
```

```python
coor_df = pd.read_csv('./prepped_data/coor_data_ver_1.csv')
print(coor_df.shape)
coor_df.head()
```

```python
merge_df['구'] = merge_df['지번주소'].apply(lambda x: x.split(' ')[1])
merge_df['동'] = merge_df['지번주소'].apply(lambda x: x.split(' ')[2])
merge_df['로'] = merge_df['도로명주소'].apply(lambda x: x.split(' ')[2])
```

```python
merge_df['동'].nunique()
```

```python
merge_df['동'].unique()
```

```python
merge_df.drop(columns=['로', '로2'], inplace=True)
```

```python
merge_df.head()
```

```python
merge_coor_df = merge_df.merge(coor_df, on=['도로명주소'])
print(merge_coor_df.shape)
merge_coor_df.head()
```

```python
%%time
merge_coor_df.to_csv('./prepped_data/main_data_ver_7.csv', index=False)
```
